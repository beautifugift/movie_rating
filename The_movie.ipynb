Convert the following content ratings to "UNRATED": TV-14, TV-G, TV-MA, TV-Y, TV-Y7.

Calculate the average imdb_score for each genre, but only include genres with at least 10 movies
Option 1: manually create a list of relevant genres, then filter using that list

```python
genresCount = df.groupby('genres').count().movie_title
genresCount >= 10

rel_genres = ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Drama', 'Horror', 'Mystery']
df[df.genres.isin(rel_genres)].groupby('genres').imdb_score.mean()
```

Calculate the average imdb_score for each genre, but only include genres with at least 10 movies
Option 1: manually create a list of relevant genres, then filter using that list

```python
genresCount = df.groupby('genres').count().movie_title
genresCount >= 10

rel_genres = ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Drama', 'Horror', 'Mystery']
df[df.genres.isin(rel_genres)].groupby('genres').movie_facebook_likes.mean()
```

Option 2: automatically create a list of relevant genres by saving the value_counts and then filtering


```python
rel_genre = df.genres.value_counts()
rel_genre = rel_genre[rel_genre >=10].index
df[df.genres.isin(rel_genre)].groupby('genres').imdb_score.mean()
```

```python
rel_genre = df.genres.value_counts()
rel_genre = rel_genre[rel_genre >=10].index
df[df.genres.isin(rel_genre)].groupby('genres').movie_facebook_likes.mean()
```

Option 3: calculate the average imdb_score/movie_facebook_likes for all genres, then filter using a boolean Series

```python
boolean_filter = df.groupby('genres').count().movie_title >= 10
df.groupby('genres').imdb_score.mean()[boolean_filter]
```

```python
boolean_filter = df.groupby('genres').count().movie_title >= 10
df.groupby('genres').movie_facebook_likes.mean()[boolean_filter]
```
